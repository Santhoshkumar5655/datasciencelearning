```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/string_processing_dataset.csv")
```

```python
df
```

```python
df["email"] = df["email"].str.lower() #we are converting into the lower case
df
```

```python
#convert cities into uppercase
df["city"] = df["city"].str.upper()
df
```

```python
#converting customer_name column into title
df["customer_name"] = df["customer_name"].str.title()
df
```

```python
df["name_length"] = df["customer_name"].str.len()
df
```

```python
df
```

```python
# regex = false tella pandas to treat the value as normal text and not as a special pattern
```

```python
df["purchase_amount"] = df["purchase_amount"].str.replace("Rs ", "", regex= False)
df
```

```python
df["purchase_amount"] = df["purchase_amount"].str.replace("$", "", regex= False)
df
```

```python
df["purchase_amount"] = df["purchase_amount"].str.replace(",", "", regex= False)
df
```

```python
df["is_mail"] = df["email"].str.contains("gmail")
df
```

```python
df["startswithv"] = df["city"].str.startswith("V")
df
```

```python
df["email_parts"] = df["email"].str.split("@").str[0]
df
```

```python
df["email_parts_combined"] = df["email"].str.split("@")
df
# chevurisanthosh@gmail.com = [chevurisanthosh, gmail.com]
```

```python
df["city_code"] = df["city"].str[:3]
df
```

```python
df["is_numeric"] = df["phone"].astype(str).str.isnumeric()
df
```

```python
df["product_code"] = df["product_code"].str.removeprefix("PRD-")
df
```

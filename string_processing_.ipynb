```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/bank_customer_data.csv")
df
```

```python
# Convert all customer names into proper format (Title Case).
df["customer_name"] = df["customer_name"].str.title()
df
# Convert all email addresses into lowercase.
df["email"] = df["email"].str.lower()
df
# Convert all bank names into uppercase.
df["bank_name"] = df["bank_name"].str.upper()
df
# Convert all account types into uppercase.
df["account_type"] = df["account_type"].str.upper()
df
# Clean the balance column by removing:
df["balance"] = df["balance"].str.replace("Rs ","", regex=False)
df["balance"] = df["balance"].str.replace("$","", regex=False)
df["balance"] = df["balance"].str.replace(",","", regex=False)
# Convert the cleaned balance column into integer datatype.
df["balance"] = df['balance'].astype(int)
df
# Create a new column called username by extracting the username from the email address.
df["user_name"] = df["email"].str.split("@").str[0]
df
df["domain"] = df["email"].str.split("@").str[1]
df
df["branch_code"] = df["branch_city"].str[:3]
df
# Remove the prefix ACC- from the account number column.
df["account_number"] = df["account_number"].str.removeprefix("ACC-")
df
```

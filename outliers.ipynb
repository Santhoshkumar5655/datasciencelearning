```python
import numpy as np
data = [2,4,5,6,7,8,9,100]
#step1 Q1, Q3
Q1 = np.percentile(data,25)
Q3 = np.percentile(data, 75)
print(Q1,Q3)
IQR = Q3 -Q1
print(IQR)
lower = Q1-1.5*IQR
upper = Q3+1.5*IQR
print(lower)
print(upper)
outliers = []
for i in data:
  if i<lower or i>upper:
    outliers.append(i)
print(outliers)
cleaned_outliers = []
for i in data:
  if lower<=i<=upper:
    cleaned_outliers.append(i)
print(cleaned_outliers)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/student_marks_outlier_dataset.csv")
```

```python
df
```

```python
#central tendency
print("mean", df["Marks"].mean())
print("median", df["Marks"].median())
print("mode", df["Marks"].mode())
```

```python
# dispersions
print("variance",df["Marks"].var())
print("standard deviation", df["Marks"].std())
```

```python
#quartiles
q1 = df["Marks"].quantile(0.25)
q2 = df["Marks"].quantile(0.50)
q3 = df["Marks"].quantile(0.75)
print("quartile1", q1)
print("quartile2", q2)
print("quartile3", q3)
```

```python
skewness = df["Marks"].skew()
print(skewness)
if skewness>0:
  print("postively skewed")
elif skewness<0:
  print("negatively skewed")
else:
  print("symteric distribution")
```

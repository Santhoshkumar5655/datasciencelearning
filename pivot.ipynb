```python
import pandas as pd
```

```python
#a pivot table is used to summarize data
df = pd.DataFrame({
    "City": ["Vijayawada", "Hyderabad", "Vijayawada", "Hyderabad", "Vijayawada"],
    "Product": ["Laptop", "Laptop", "Mouse", "Mouse", "Laptop"],
    "Sales": [50000, 60000, 2000, 3000, 40000]
})
```

```python
df
```

```python
df.groupby(["City", "Product"])["Sales"].sum()
```

```python
pd.pivot_table(
    df,
    values = "Sales",
    index= "City",
    columns="Product",
    aggfunc="sum"
)
```

```python
pd.pivot_table(
    df,
    values ="Sales",
    index= "City",
    aggfunc="sum"

)
df.groupby("City")["sales"].sum()
```

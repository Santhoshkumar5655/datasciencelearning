```python
# # one hot encoding
# # one hot encoding is technique used to
# # convert text data into numaerical data by creating separate columns for each
# # category and filling thems with os and 1s
# # 1 = category presnt
# # 0 = category not present
# red = [1 0 0] 1 0 0
# green = [010] 0 1 0
# blue  = [001] 0 0 1
```

image.png

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df = sns.load_dataset("tips")
X = df[["sex"]]
encoder = OneHotEncoder(sparse_output = False)
encoded = encoder.fit_transform(X)
print(encoded)
print(encoder.get_feature_names_out())
encoded_df = pd.DataFrame(
    encoded,
    columns = encoder.get_feature_names_out()

)
encoded_df
d = df.drop("sex", axis =1)
final_df = pd.concat([d,encoded_df], axis = 1)
final_df
```

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df = sns.load_dataset("diamonds")
df
print(df.describe())
print(df.shape)
print(df.columns)
cut_data = df[["cut"]]
encoder = OneHotEncoder(sparse_output=False)
encoded_cut = encoder.fit_transform(cut_data)
print(encoded_cut)
cut_df = pd.DataFrame(
   encoded_cut,
   columns= encoder.get_feature_names_out(["cut"])
)
cut_df
color_data = df[["color"]]
encoder = OneHotEncoder(sparse_output=False)
encoded_color = encoder.fit_transform(color_data )
print(encoded_color)
color_df = pd.DataFrame(
   encoded_color,
   columns= encoder.get_feature_names_out(["color"])
)
color_df
clarity_data = df[["clarity"]]
encoder = OneHotEncoder(sparse_output=False)
clarrity_encoded = encoder.fit_transform(clarity_data )
print(clarrity_encoded )
clarity_df = pd.DataFrame(
    clarrity_encoded,
   columns= encoder.get_feature_names_out(["clarity"])
)
clarity_df
final_df = pd.concat([df,cut_df,color_df,clarity_df], axis=1)
```

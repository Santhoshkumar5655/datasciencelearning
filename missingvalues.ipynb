```python
import seaborn as sns
df = sns.load_dataset('titanic')
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["age"]
```

```python
df.columns
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
print(df["age"])
```

```python
df
```

```python
df[["Age_Mean", "age"]]
```

```python
df["Age_Median"] = df["age"].fillna(df["age"].median())
```

```python
df[["Age_Median", "age"]]
```

```python
df["Mode"] = df["age"].fillna(df["age"].mode()[0])
print(df["Mode"])

```

```python
#converting data column
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
#sort the values
dfd = df.sort_values("Date")
dfd
```

```python
#set date as index
dfd.set_index("Date", inplace= True)
dfd
```

```python
dfd
```

```python
#so we can create missing dates too
dfd = dfd.asfreq("D")
dfd
# code-meaning
# d - daily
# w - weekly
# m-monthly
# y-yearly
# h-hourly
```

```python
#handling missing values
# day1 = 100
# day2 =
# day3 =
# day4 = 160
# 60/3 = 20
# interpolation means filling missing Values by using
# the values before and after them and estimating
# what the missing value should be
dfd["Stock_Price"] = dfd["Stock_Price"].interpolate()
dfd["Volume"] = dfd["Volume"].interpolate()
dfd
```

```python
#forwardfill replaces a missing value with the previous avilable value
# (ffill)
#backwardpill replcaes a misiing value with the next available valuebfill
```

```python
dfd
```

```python
#convert date features
dfd["Year"] = dfd.index.year
dfd["Month"] = dfd.index.month
dfd["Day"] = dfd.index.day
dfd
dfd["dayofweek"] = dfd.index.dayofweek
dfd
# m - 0
# t-1
# w-2
# t-3
# f-4
# s-5
# su-6
```

```python
dfd["weekend"] = (dfd.index.dayofweek>=5).astype(int)
dfd
```

```python
#lag feature
# a lag feature means  use past values to predict future values
dfd["lag1"] = dfd["Stock_Price"].shift(1)
dfd
dfd["lag2"] = dfd["Stock_Price"].shift(2)
dfd
dfd["moving_average"] = dfd["Stock_Price"].rolling(3).mean()
dfd
# 1 1000
# 2 1110
# 3 2500
dfd["max_roll"] = dfd["Stock_Price"].rolling(3).max()
dfd
```

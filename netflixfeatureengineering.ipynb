```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/netflix_titles.csv")
```

```python
df.info()
```

```python
#missing values check
df.isnull().sum()
```

```python
df["director"] = df["director"].fillna("Unknown")
df["country"] = df["country"].fillna("UnKnown")
df["cast"] = df["cast"].fillna("Unknown")
df["date_added"] = df["date_added"].fillna("January 1, 2020")
df["rating"] = df["rating"].fillna("Not Rated")
df["duration"] = df["duration"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#date features
df["date_added"] = pd.to_datetime(
    df["date_added"], format="mixed"
)
df.info()
df["ADDED_YEAR"] = df["date_added"].dt.year
df["Added_month"] = df["date_added"].dt.month
df["Added_day"] = df["date_added"].dt.day
df["addedweekday"] = df["date_added"].dt.day_name()
#q1 = jan-march
#q2 = april -jun
#q3= july -sept
#q4 = oct-dec
df["AddedQuarter"] = df["date_added"].dt.quarter
df.head(1)
```

```python
df
#current year - release_year
# 2026- release _year
# 2026-2020 = 5
```

```python
df.columns
```

```python
df["ContentAge"] = (2026-df["release_year"])
df[["release_year","ContentAge"]].head(2)
df["YearstoNetflix"] = (df["ADDED_YEAR"]-df["release_year"])
df
```

```python
df["Title_length"] = df["title"].str.len()
df["Title_length"]
```

```python
df["title"]
```

```python
# TitleWordCount Count words in title
df["titlewordcount"] = df["title"].str.split().str.len()
df[["title","titlewordcount"]].head(1)
```

```python
df["LongTitle"] = (df["titlewordcount"]>3).astype(int)
df[["title","titlewordcount","LongTitle"]].head(1)
```

```python
df["descriptionlength"] = df["description"].str.len()
df["DescriptionWordCount"] = df["description"].str.split().str.len()
```

```python
df["actor_count"] = df["cast"].str.count(",") +1
# jagadeesh = 0+1
# jagadeesh,govind,mahesh = 2+1
# lahari,vara,swathi, divya = 3+1
df["Country_count"] = df["country"].str.count(",")+1
df
```

```python
df.columns
```

```python
df["generecount"] = df["listed_in"].str.count(",")+1
df["generecount"]
```

```python
df["multigenre"] = (
    df["generecount"]>1
).astype(int)
df["multigenre"]
```

```python
df["type"]
```

```python
df["ismovie"] = (df["type"]=="Movie").astype(int)
df["istvshows"] = (df["type"]=="TV Show").astype(int)
```

```python
df["recet_content"] = (df["release_year"]>=2021).astype(int)
df["classic_content"] = (df["release_year"]<=2021).astype(int)
```
